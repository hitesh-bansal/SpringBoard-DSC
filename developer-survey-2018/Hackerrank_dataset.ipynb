## As we can see there are some null values for profession, so we have to drop these rows

```python
df = df_professional.dropna(subset=['q8JobLevel'])
```

```python
df_woman = df.loc[df.q3Gender == 'Female']
```

```python
df_male = df.loc[df.q3Gender == 'Male']
```

## Professional Age when they started coding

```python
cols = ['q1AgeBeginCoding','q2Age']
```

```python
df_woman_age = df_woman[cols]
```

```python
df_male_age = df_male[cols]
```

```python
df_woman_age.head()
```

```python
df_male_age.info()
```

```python
df_woman_age.isnull().values.any()
```

```python
df_male_age.isnull().values.any()
```

```python
indexNames = df_woman_age[ (df_woman_age['q1AgeBeginCoding'] == '#NULL!') | (df_woman_age['q2Age'] == '#NULL!') ].index
df_woman_age.drop(indexNames , inplace=True)
```

```python
indexNames = df_male_age[ (df_male_age['q1AgeBeginCoding'] == '#NULL!') | (df_male_age['q2Age'] == '#NULL!') ].index
df_male_age.drop(indexNames , inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
WomanAgeBeginCodingcounts = df_woman_age['q1AgeBeginCoding'].value_counts()
```

```python
MaleAgeBeginCodingcounts = df_male_age['q1AgeBeginCoding'].value_counts()
```

```python
WomanAgeBeginCodingcounts
```

```python
MaleAgeBeginCodingcounts
```

```python
fig = plt.figure(figsize=(6,6)) # define plot area
ax = fig.gca() # define axis    
WomanAgeBeginCodingcounts.plot.bar(ax = ax) # Use the plot.bar method on the counts data frame
ax.set_xlabel('Age of woman professional when they started coding.') # Set text for the x axis
ax.set_ylabel('Count the number of woman professional')# Set text for y axis
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.gca()
MaleAgeBeginCodingcounts.plot.bar(ax = ax) 
ax.set_xlabel('Age of male professional when they started coding.') 
ax.set_ylabel('Count the number of male professional')
```

```python
WomanAgecounts = df_woman_age['q2Age'].value_counts()
```
