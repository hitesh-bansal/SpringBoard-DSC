### No. of Data per Activity

```python
plt.title('No of Data per Activity', fontsize=15)
sns.countplot(df.Activity)
plt.xticks(rotation=75)
plt.show()
```

```python
df.plot(x = 'Activity',y = 'tBodyAcc-mean()-X', figsize= (15,15), grid=True, title="tBodyAcc-mean()-X")
```

```python
df.plot(x = 'Activity',y = ['tBodyAcc-mean()-X','tBodyAcc-mean()-Y','tBodyAcc-mean()-Z'], figsize= (15,15), grid=True, title="tBodyAcc-mean()-X VS tBodyAcc-mean()-Y VS tBodyAcc-mean()-Z")
```

```python
plt.figure(figsize=(14,7))
plt.subplot(2,2,1)
plt.title('Stationary Activities')
sns.distplot(df_SITTING['angle(X,gravityMean)'],hist = False,color = 'r', label = 'Sitting')
sns.distplot(df_STANDING['angle(X,gravityMean)'],hist = False, color = 'm', label = 'Standing')
sns.distplot(df_LAYING['angle(X,gravityMean)'],hist = False, color = 'c', label = 'Laying')
plt.legend(loc='center')

plt.subplot(2,2,2)
plt.title('Moving Activities')
sns.distplot(df_WALKING['angle(X,gravityMean)'],hist = False,color = 'r',label = 'Walking')
sns.distplot(df_WALKING_UPSTAIRS['angle(X,gravityMean)'],hist = False,color = 'm',label = 'walking upstair')
sns.distplot(df_WALKING_DOWNSTAIRS['angle(X,gravityMean)'],hist = False,color = 'c',label = 'walking downstair')
plt.legend(loc='center right')

plt.tight_layout()
plt.show()
```

```python
sns.set_palette("Set1", desat=0.80)
facetgrid = sns.FacetGrid(df, hue='Activity', size=6,aspect=2)
facetgrid.map(sns.distplot,'tBodyAcc-std()-X', hist=False)\
    .add_legend()
plt.annotate("Stationary Activities", xy=(-0.956,17), xytext=(-0.9, 23), size=20,\
            va='center', ha='left',\
            arrowprops=dict(arrowstyle="simple",connectionstyle="arc3,rad=0.1"))

plt.annotate("Moving Activities", xy=(0,3), xytext=(0.2, 9), size=20,\
            va='center', ha='left',\
            arrowprops=dict(arrowstyle="simple",connectionstyle="arc3,rad=0.1"))
plt.show()
```
